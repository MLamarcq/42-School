```python
def convert_to_string(element) :
    if not isinstance(element, str) : 
        return str(element)
    return element
```

```python
for col in france_cleaned.columns :
    print(col)
    france_cleaned[col] = france_cleaned[col].astype(str)
    france_cleaned.loc[:, col] = france_cleaned[col].apply(convert_to_string)
    # print(f"le resultat = {france_cleaned[col]}")
```

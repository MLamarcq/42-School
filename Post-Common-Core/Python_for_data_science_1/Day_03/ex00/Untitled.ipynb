```python
def convert_str(element) :
    if isinstance(element, str) :
        if 'M' in element :
            return float(element.replace('M', '').replace(',','')) * 1e6
        elif 'B' in element :
            return float(element.replace('B', '').replace(',','')) * 1e9
        else :
            return float(element.replace(',',''))
    return np.nan
```

```python
grouped.loc[:, '1800']
```

```python
for col in grouped.iloc[:, 1:] :
	grouped.loc[:, col] = grouped[col].apply(convert_str)
```
